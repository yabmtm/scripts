```r
library(rayshader)
library(ggplot2)
library(sf)
library(viridis)
library(readr)
library(RcppCNPy)
library(rgl)

tica1 <- npyLoad('tica/tica1.npy')
tica2 <- npyLoad('tica/tica2.npy')

a = data.frame(x=tica1, y=tica2)

tica_1_2 = ggplot(a, aes(x=x, y=y)) +
  geom_hex(bins = 3, size = 0.5, color = "black") +
  scale_fill_viridis_c(option = "C")

tica_1_2
```

```r
plot_gg(tica_1_2, multicore = TRUE, width = 8, height = 8, scale = 2, 
          zoom = 0.6, phi = 60,
          background = "#afceff",shadowcolor = "#3a4f70")

# render_movie("tica_1_2.mp4", frames = 480)
```
